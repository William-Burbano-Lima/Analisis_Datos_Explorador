# Talento Tech Bootcamp Analisis de Datos

Explorador

Ejecutor Técnico: William Burbano Lima



**Plan de Clase Virtual de 3 Horas: Análisis de Datos**

**Objetivo del Curso:**

Proveer a los estudiantes de Talento Tech con un entendimiento fundamental de los conceptos de estadística descriptiva y su aplicación en el análisis de datos mediante el uso de los datasets Titanic e IRIS.

**Agenda 26-06-2024**

Hora 1: Introducción a la Estadística Descriptiva y Análisis Exploratorio de Datos (EDA)

00:00 - 00:15 - Bienvenida y Presentación del Curso

    * Objetivos del curso.
    * Resumen de la agenda.

00:15 - 00:30 - Conceptos Básicos de Estadística Descriptiva

    Definición y explicación de:
        * Media (Promedio)
        * Mediana
        * Moda
        * Desviación Estándar
    Ejemplos sencillos para ilustrar cada concepto.

00:30 - 00:45 - Importancia del Análisis Exploratorio de Datos (EDA)

    * Definición y objetivos del EDA.
    * Pasos comunes en el EDA.
    * Ejemplos de preguntas que el EDA puede responder.

00:45 - 01:00 - Primer Ejercicio: Cálculo Manual de Estadísticas Básicas

    * Utilizando un pequeño conjunto de datos ficticios, los estudiantes calcularán manualmente la media, mediana, moda y desviación estándar.
    * Discusión de los resultados y comparación con los conceptos teóricos.

Hora 2: Análisis Exploratorio del Dataset Titanic

01:00 - 01:15 - Introducción al Dataset Titanic

    * Descripción del dataset: columnas y significado de cada una.
    * Preguntas que podemos responder con este dataset.

01:15 - 01:45 - Análisis Descriptivo del Dataset Titanic en Excel

    * Carga del dataset Titanic en Excel.
    * Exploración inicial de los datos.
    * Cálculo de estadísticas básicas (media, mediana, moda, desviación estándar) para la columna Age.
    * Visualización de datos con gráficos de barras y histogramas.

01:45 - 02:00 - Discusión de Resultados y Conclusiones

    * Interpretación de las estadísticas calculadas.
    * Discusión sobre la diversidad de edades en el Titanic y su relación con la supervivencia.
    * Preguntas y respuestas.

Hora 3: Aplicación de Estadística Descriptiva en el Dataset IRIS

02:00 - 02:15 - Introducción al Dataset IRIS

    * Descripción del dataset: columnas y significado de cada una.
    * Preguntas que podemos responder con este dataset.

02:15 - 02:45 - Análisis Descriptivo del Dataset IRIS en Excel

    * Carga del dataset IRIS en Excel.
    * Exploración inicial de los datos.
    * Cálculo de estadísticas básicas (media, mediana, moda, desviación estándar) para las columnas sepal_length, sepal_width, petal_length, petal_width.
    * Comparación de las estadísticas entre las especies (setosa, versicolor, virginica).

02:45 - 03:00 - Discusión de Resultados y Conclusiones Finales

  *  Interpretación de las estadísticas calculadas para cada especie.
  * Conclusiones sobre las diferencias entre las especies en base a las medidas descriptivas.
  * Preguntas y respuestas.
   * Evaluación de la clase y comentarios finales.

Recursos y Materiales Necesarios

    Datasets:
        Titanic: Descargar
        IRIS: Disponible en la mayoría de las bibliotecas de datos y en múltiples sitios web educativos.

    Software:
        Excel para la carga y análisis de los datasets.
        Opcional: Herramientas adicionales como Google Sheets o cualquier otra herramienta de hoja de cálculo.

    Material Didáctico:
        Presentaciones en PowerPoint o Google Slides que cubren los conceptos teóricos y los pasos del EDA.
        Guías de ejercicios con ejemplos prácticos para que los estudiantes sigan durante la clase.

Evaluación y Seguimiento

   *  Participación Activa: Evaluar la participación y comprensión a través de preguntas y respuestas durante la clase.
   *  Ejercicios Prácticos: Revisar los cálculos manuales y las conclusiones de los ejercicios en Excel.
   *  Comentarios y Retroalimentación: Al final de la clase, solicitar comentarios para mejorar futuras sesiones.
