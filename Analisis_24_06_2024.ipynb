# TALENTO TECH

**Plan de Clase Virtual para Bootcamp de Análisis de Datos**

EJECUTOR TECNICO : William Burbano Lima

fecha: 24-06-2024

Objetivo General:

Introducir a los estudiantes a las técnicas fundamentales de análisis de datos, incluyendo la limpieza de datos, el análisis exploratorio y la visualización de datos utilizando Excel.
Objetivos Específicos:

    * Comprender los conceptos básicos del análisis de datos y su importancia.
    * Aprender a manejar y limpiar datos en Excel.
    * Realizar análisis exploratorios y visualizaciones básicas en Excel.
    * Aplicar técnicas de análisis de datos a un caso práctico.

Estructura de la Clase:
Hora 1: Introducción y Conceptos Básicos

1.1 Bienvenida y Presentación del Instructor (10 minutos)

    Introducción del instructor y objetivos de la clase.
    Explicación de la importancia del análisis de datos.

1.2 Conceptos Básicos del Análisis de Datos (20 minutos)

    Definición de análisis de datos.
    Tipos de datos (estructurados vs no estructurados).
    Ciclo de vida del análisis de datos: recopilación, limpieza, análisis, visualización e interpretación.

1.3 Herramientas y Software para el Análisis de Datos (10 minutos)

    Breve introducción a las herramientas más comunes (Excel, Python, R).
    Enfoque en Excel para el curso.

1.4 Preguntas y Respuestas (5 minutos)
Hora 2: Manejo y Limpieza de Datos en Excel

2.1 Importación de Datos en Excel (15 minutos)

    Cómo importar datos desde diferentes fuentes (CSV, TXT, etc.).

2.2 Estrategias de Limpieza de Datos (25 minutos)

    Eliminación de duplicados.
    Manejo de valores faltantes.
    Normalización de datos.
    Formateo de celdas y tipos de datos.

2.3 Ejercicio Práctico: Limpieza de Datos (15 minutos)

    Proporcionar un archivo de datos desordenado.
    Guía paso a paso para limpiar los datos en Excel.

2.4 Preguntas y Respuestas (5 minutos)
Hora 3: Análisis Exploratorio y Visualización

3.1 Análisis Exploratorio de Datos (20 minutos)

    Descripción estadística de datos.
    Uso de funciones básicas de Excel (PROMEDIO, MEDIANA, DESVEST).

3.2 Visualización de Datos (20 minutos)

    Introducción a los gráficos en Excel.
    Cómo crear gráficos de barras, líneas y dispersión.
    Interpretación de gráficos.

3.3 Ejercicio Práctico: Análisis y Visualización (15 minutos)

    Proporcionar un conjunto de datos estructurados.
    Realizar un análisis exploratorio y crear visualizaciones.

3.4 Preguntas y Respuestas y Cierre (5 minutos)

    Resolución de dudas finales.
    Resumen de los conceptos aprendidos.
    Información sobre recursos adicionales y próximos pasos.

Materiales Necesarios:

    Computadora con Microsoft Excel instalado.
    Conexión a Internet.
    Archivos de datos para ejercicios prácticos (CSV, TXT).

Actividades y Ejercicios:

    Ejercicio de Limpieza de Datos:
        Proporcionar un archivo CSV desordenado.
        Instrucciones para limpiar y formatear los datos.

    Análisis Exploratorio:
        Datos de ejemplo sobre ventas de una tienda.
        Cálculo de estadísticas descriptivas.

    Visualización de Datos:
        Crear gráficos de ventas mensuales.
        Interpretar las tendencias a partir de los gráficos.

Recursos Adicionales:

    Enlaces a tutoriales en línea sobre Excel.
    Documentación y guías de usuario para Excel.
    Libros y artículos recomendados sobre análisis de datos.

Evaluación:

    Participación activa durante la clase.
    Completar los ejercicios prácticos.
    Realizar una breve autoevaluación al final de la clase sobre los conceptos aprendidos.
