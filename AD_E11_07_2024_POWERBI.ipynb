- **Análisis Bivariante:**
  - Análisis de dos variables para entender la relación entre ellas.
  - Ejemplos: Relación entre clase y supervivencia, sexo y supervivencia.

### Ejemplo de Actividad en Power BI:

#### Importación y Preparación de Datos:
1. **Importar datos:**
   - Archivo > Obtener datos > Archivo > Excel > Seleccionar el archivo del Titanic.

   - Crear columnas calculadas, como `AgeGroup` basada en intervalos de edad.

#### Creación de Visualizaciones:
1. **Gráficos Básicos:**
   - Crear un gráfico de barras para `Survived` y `Pclass`.
   - Crear un histograma para `Age`.
2. **Gráficos Avanzados:**
   - Crear un gráfico de dispersión para analizar la relación entre `Fare` y `Age`.
   - Crear un mapa para visualizar el puerto de embarque (`Embarked`).

#### Creación de Dashboards:
1. **Combinar visualizaciones:**

2. **Interactividad:**
