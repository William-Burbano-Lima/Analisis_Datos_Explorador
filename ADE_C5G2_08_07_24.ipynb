#### Objetivo:

### Hora 1: Instalación e Introducción a Power BI

**1.1 Introducción a Power BI (15 minutos)**
- Explicación de qué es Power BI y sus beneficios para la visualización de datos.
- Diferencias entre Power BI y otras herramientas de visualización de datos.

**1.2 Instalación de Power BI (15 minutos)**
- Guía paso a paso para descargar e instalar Power BI Desktop.
  ```plaintext
  1. Ir a la página oficial de Power BI (https://powerbi.microsoft.com/desktop/)
  2. Descargar Power BI Desktop.
  3. Ejecutar el instalador y seguir las instrucciones.
  ```

**1.3 Primeros Pasos en Power BI (30 minutos)**
- Importar el dataset `titanic.csv` a Power BI.
  ```plaintext
  1. Abrir Power BI Desktop.
  2. Ir a `Inicio -> Obtener datos -> Texto/CSV`.
  3. Seleccionar el archivo `titanic.csv` y cargar los datos.
  ```
- Navegación básica en Power BI: panel de campos, visualizaciones y filtros.

- Limpieza de datos: eliminar filas nulas, transformar datos.
  ```plaintext
  1. Seleccionar la tabla en el panel de campos.
  2. Ir a `Transformar datos`.
  3. Realizar operaciones de limpieza (eliminar valores nulos, cambiar tipos de datos, etc.).
  ```
