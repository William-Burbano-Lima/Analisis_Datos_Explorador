**Talento Tech **
**Bootcamp de Análisis de Datos**

Nivel Explorador

Fecha: junio 22 2024

Duración: 5 Horas

Ejecutor técnico: William Burbano Lima

**Objetivos del Bootcamp:**

  * Introducción a los conceptos básicos del análisis de datos.
  * Manejo inicial de archivos de datos y bases de datos.
  * Contextualización con ejemplos prácticos del mundo real.
  * Ejercicios prácticos utilizando Excel.
    Implementación de una metodología ágil en el aprendizaje.

**Agenda del Bootcamp**:

Hora 1: Introducción al Análisis de Datos y Metodología Ágil

  * Bienvenida y Presentación
  * Introducción del instructor
  * Objetivos del bootcamp
  * Agenda del día

  **Conceptos Básicos del Análisis de Datos**
    * Definición y importancia
    * Tipos de datos: estructurados y no estructurados
    * Aplicaciones del análisis de datos en diversas industrias

  **Actividad Interactiva: Reflexión sobre Casos de Uso**
        * Discusión en grupos pequeños
        * Presentación de ideas

Hora 2: Manejo Inicial de Archivos de Datos

    * Introducción a Archivos CSV y Excel
    * Formatos de archivo y su importancia
    * Herramientas para trabajar con estos archivos

    Importación y Limpieza de Datos en Excel
        Importar un archivo CSV
        Eliminar duplicados
        Manejar valores faltantes

    Discusión y Feedback

Hora 3: Contextualización con Ejemplos Prácticos del Mundo Real

    Casos de Estudio en Diferentes Industrias
        Marketing
        Salud
        Finanzas

    Análisis de un Caso de Estudio
        Discusión en grupos
        Identificación de problemas
        Propuestas de solución

    Presentación de Resultados

Hora 4: Ejercicios Prácticos Usando Excel

    Análisis Descriptivo en Excel
        Creación de gráficos básicos
        Tablas dinámicas

    Aplicación de Funciones Estadísticas
        Promedio
        Mediana
        Desviación estándar

    Ejercicio Práctico: Mini Proyecto en Excel
        Análisis de un conjunto de datos
        Generación de informes visuales y descriptivos
        Presentación de resultados

Hora 5: Metodología Ágil y Retroalimentación

    Mini Sprint Ágil: Iteración y Mejora

        Presentación de los proyectos mejorados

        Sesión de preguntas y respuestas

Recursos Necesarios:

    Computadoras con Excel instalado
    Conexión a Internet
    Proyector y pantalla para presentaciones
    Archivos de datos para ejercicios prácticos
    Material impreso o digital de apoyo (presentaciones, guías de ejercicios)

Ejercicios y Actividades:

    * Actividad Interactiva: Reflexión sobre cómo los datos se utilizan en diferentes industrias. Discusión en grupos pequeños y presentación de ideas.
    * Ejercicio Práctico 1: Importar datos desde un archivo CSV a Excel y realizar limpieza de datos (manejo de valores faltantes y eliminación de duplicados).
    * Ejercicio Práctico 2: Análisis de un caso de estudio proporcionado. Identificación de problemas y propuestas de solución.
    * Ejercicio Práctico 3: Creación de gráficos básicos y tablas dinámicas en Excel. Aplicación de funciones estadísticas.
    * Mini Proyecto: Análisis de un conjunto de datos utilizando técnicas aprendidas. Generación de informes visuales y descriptivos. Presentación de resultados.

Evaluación del Bootcamp:

    Participación en actividades y ejercicios prácticos.
    Presentación de resultados de ejercicios prácticos y mini proyecto.
    Feedback continuo y retroalimentación al final del bootcamp.
