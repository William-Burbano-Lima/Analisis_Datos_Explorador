## Resumen de Conceptos Básicos de Análisis de Datos

### 1. ¿Qué es el Análisis de Datos?

El análisis de datos es el proceso de inspeccionar, limpiar y modelar datos con el objetivo de descubrir información útil, llegar a conclusiones y apoyar la toma de decisiones. Es fundamental para transformar datos en información valiosa y comprensible.

### 2. Tipos de Datos

- **Datos Cuantitativos**: Números que pueden medirse y contarse (e.g., goles en un partido).
  - **Discretos**: Valores contables (e.g., número de goles).
  - **Continuos**: Valores dentro de un rango (e.g., tiempo de posesión).
- **Datos Cualitativos**: Descripciones o categorías (e.g., nombres de equipos).
  - **Nominales**: Categorías sin orden (e.g., colores de camisetas).
  - **Ordinales**: Categorías con un orden lógico (e.g., posiciones en la liga).

### 3. Análisis Exploratorio de Datos (EDA)

El EDA es una etapa crucial del análisis de datos donde se utilizan técnicas estadísticas y de visualización para comprender las características y relaciones de los datos. Los principales pasos incluyen:

- **Inspección Inicial**: Revisión de la estructura de los datos, tipos de variables y valores faltantes.
- **Estadísticas Descriptivas**: Cálculo de medidas como media, mediana, moda, rango, desviación estándar, etc.
- **Visualización de Datos**: Uso de gráficos para identificar patrones y relaciones. Los tipos comunes de gráficos incluyen:
  - **Histogramas**: Para mostrar la distribución de una variable continua.
  - **Gráficos de Barras**: Para comparar diferentes categorías.
  - **Gráficos de Líneas**: Para mostrar tendencias a lo largo del tiempo.
  - **Gráficos de Dispersión**: Para visualizar la relación entre dos variables.

### 4. Estadísticas Descriptivas

- **Medidas de Tendencia Central**:
  - **Media**: El promedio de un conjunto de datos.
  - **Mediana**: El valor central en un conjunto de datos ordenados.
  - **Moda**: El valor que ocurre con mayor frecuencia.
- **Medidas de Dispersión**:
  - **Rango**: Diferencia entre el valor máximo y mínimo.
  - **Varianza**: Promedio de las desviaciones al cuadrado respecto a la media.
  - **Desviación Estándar**: Raíz cuadrada de la varianza, indica la dispersión de los datos respecto a la media.

### 5. Tipos de Análisis de Datos

- **Análisis Descriptivo**: Describe las características básicas de los datos mediante estadísticas y gráficos.
- **Análisis Diagnóstico**: Identifica causas de patrones y comportamientos observados en los datos.
- **Análisis Predictivo**: Utiliza datos históricos para predecir futuros resultados mediante modelos estadísticos y algoritmos de machine learning.
- **Análisis Prescriptivo**: Ofrece recomendaciones basadas en los resultados del análisis predictivo para la toma de decisiones.

### 6. Visualización de Datos en Excel

Excel ofrece diversas herramientas para crear gráficos y visualizar datos:

- **Histogramas**: Para mostrar la distribución de datos continuos.
- **Gráficos de Barras**: Para comparar diferentes categorías.
- **Gráficos de Líneas**: Para mostrar tendencias en datos temporales.
- **Gráficos de Dispersión**: Para examinar relaciones entre dos variables.

### 7. Importancia del EDA

El EDA ayuda a:

- **Comprender la estructura y propiedades de los datos**: Identificar valores atípicos, datos faltantes y relaciones entre variables.
- **Guiar el análisis posterior**: Informar sobre qué técnicas estadísticas y modelos pueden ser aplicables.
- **Comunicar hallazgos**: Facilitar la presentación de resultados de manera clara y visualmente atractiva.
