**3.1 Uso de Funciones Estadísticas Complejas (30 minutos)**
- **Percentiles y Cuartiles**:
  - Cálculo de percentiles y cuartiles para las variables `Age` y `Fare`.
  ```excel
  =PERCENTIL.EXC(rango_de_datos, percentil)
  ```

- **Función `FREQUENCY`**:
  - Uso de `FREQUENCY` para crear distribuciones de frecuencia.
