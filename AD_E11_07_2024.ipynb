### Análisis de Datos con Power BI

#### **Paso 1: Preparación del Entorno**

1. **Instalación de Power BI Desktop:**
   - Descarga e instala Power BI Desktop desde [aquí](https://powerbi.microsoft.com/desktop/).

2. **Dataset del Titanic:**
   - Descarga el dataset del Titanic desde [Kaggle](https://www.kaggle.com/c/titanic/data) o utiliza el archivo CSV proporcionado.

#### **Paso 2: Importación de Datos**

1. **Abrir Power BI Desktop:**
   - Inicia Power BI Desktop.

2. **Importar el Dataset:**
   - Haz clic en `Obtener datos` en la cinta de opciones.
   - Selecciona `Texto/CSV` y busca el archivo del Titanic en tu computadora.
   - Haz clic en `Cargar`.

#### **Paso 3: Limpieza y Preparación de Datos**

1. **Abrir el Editor de Consultas:**
   - En la cinta de opciones, selecciona `Transformar datos` para abrir el Editor de Consultas.

2. **Limpieza de Datos:**
   - **Eliminar columnas innecesarias:**
     - Selecciona columnas que no necesitas (por ejemplo, `Cabin` si está mayormente vacío) y haz clic en `Eliminar columnas`.
   - **Manejo de valores nulos:**
     - Selecciona una columna con valores nulos (por ejemplo, `Age`).
     - En la pestaña `Transformar`, selecciona `Reemplazar valores` y reemplaza los valores nulos por la mediana o media de la columna.
   - **Conversión de tipos de datos:**
     - Asegúrate de que todas las columnas tienen el tipo de dato correcto. Puedes cambiar el tipo de dato seleccionando la columna y usando la opción `Tipo de datos` en la pestaña `Transformar`.

3. **Crear Columnas Calculadas:**
   - **Crear un grupo de edad:**
     - Selecciona la columna `Age`.
     - En la pestaña `Agregar columna`, selecciona `Columna condicional`.
     - Define las condiciones para diferentes grupos de edad (por ejemplo, `0-10`, `11-20`, etc.).

4. **Cerrar y Aplicar:**
   - Una vez que hayas terminado con las transformaciones, haz clic en `Cerrar y aplicar`.

#### **Paso 4: Creación de Visualizaciones**

1. **Gráfico de Barras: Supervivencia por Clase**
   - En la vista de `Informe`, selecciona `Gráfico de barras apiladas`.
   - Arrastra `Pclass` al eje `Eje` y `Survived` al `Valores`.
   - Configura el gráfico para mostrar la distribución de supervivientes por clase.

2. **Histograma: Distribución de Edades**
   - Selecciona `Gráfico de columnas agrupadas`.
   - Arrastra `Age` al eje `Eje` y usa una función de `Contar` para los valores.

3. **Gráfico de Dispersión: Relación entre Tarifa y Edad**
   - Selecciona `Gráfico de dispersión`.
   - Arrastra `Fare` al eje `X` y `Age` al eje `Y`.
   - Usa `Survived` como leyenda para diferenciar entre supervivientes y no supervivientes.

4. **Mapa: Puertos de Embarque**
   - Selecciona `Mapa`.
   - Arrastra `Embarked` al campo `Ubicación`.
   - Configura para mostrar la cantidad de pasajeros que embarcaron en cada puerto.

#### **Paso 5: Creación de Dashboards Interactivos**

1. **Combinar Visualizaciones:**
   - Arrastra y suelta las diferentes visualizaciones que creaste en una sola página de informe.

2. **Añadir Segmentaciones:**
   - En la cinta de opciones, selecciona `Segmentación de datos`.
   - Arrastra `Pclass` y `Sex` a las segmentaciones.
   - Configura las segmentaciones para que puedan filtrar los datos en el informe.

3. **Configurar Interactividad:**
   - Selecciona una visualización, haz clic en `Editar interacciones` en la pestaña `Formato`.
   - Configura cómo las visualizaciones interactúan entre sí (por ejemplo, filtrar o resaltar datos).

#### **Paso 6: Publicación y Compartición**

1. **Publicar Informe:**
   - Haz clic en `Publicar` en la cinta de opciones.
   - Inicia sesión en tu cuenta de Power BI y selecciona un área de trabajo para publicar el informe.

2. **Compartir Dashboard:**
   - En Power BI Service, navega al informe publicado.
   - Usa la opción de `Compartir` para enviar el informe a otros usuarios.
