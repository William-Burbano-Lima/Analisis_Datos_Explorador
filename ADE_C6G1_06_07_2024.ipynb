3. **Limpieza de Datos (8:45 AM - 9:15 AM)**
   - **Identificación de Valores Nulos:**
     - Muestra cómo filtrar y encontrar valores nulos en Excel.
   - **Imputación de Valores Nulos:**
     - Imputa los valores nulos en la columna `Age` con la mediana:
       - `=SI(ESBLANCO(B2), MEDIANA($B$2:$B$100), B2)`
     - Imputa los valores nulos en la columna `Embarked` con el modo:
       - `=SI(ESBLANCO(E2), MODA($E$2:$E$100), E2)`
   - **Eliminación de Valores Atípicos:**
     - Identifica y elimina valores atípicos utilizando gráficos y estadísticos descriptivos.

4. **Descripción Estadística (9:15 AM - 10:00 AM)**
   - **Medidas de Tendencia Central:**
     - Calcula la media: `=PROMEDIO(B:B)` (suponiendo que `Age` está en la columna B).
     - Calcula la mediana: `=MEDIANA(B:B)`.
     - Calcula la moda: `=MODA.UNIMODAL(B:B)`.
   - **Medidas de Dispersión:**
     - Calcula la desviación estándar: `=DESVEST(B:B)`.
     - Calcula los cuartiles: `=CUARTIL.INC(B:B, 1)` para el primer cuartil, `=CUARTIL.INC(B:B, 2)` para la mediana, y `=CUARTIL.INC(B:B, 3)` para el tercer cuartil.
   - **Análisis de la Distribución de los Datos:**
     - Discute la distribución de los datos basándose en las estadísticas calculadas.
