#### Objetivos de la Clase
1. Repasar y consolidar conceptos fundamentales del análisis exploratorio de datos.
2. Desarrollar habilidades avanzadas en el uso de Excel para el análisis de datos.
3. Introducir técnicas de visualización y análisis de datos, con Power BI.

#### Agenda de la Clase

**2:00 PM - 2:30 PM:** Introducción y objetivos de la clase
- Breve repaso de los conceptos básicos de análisis de datos.
- Presentación de los temas del día.

**2:30 PM - 3:30 PM:** Análisis Exploratorio con Excel usando el dataset de Titanic
- **Fórmulas de tendencia central:**
  - **Media:** `=PROMEDIO(rango_de_datos)`
  - **Mediana:** `=MEDIANA(rango_de_datos)`
  - **Moda:** `=MODA(rango_de_datos)`
  - **Paso a Paso:**
    - Selecciona las celdas con los datos que quieres analizar.
    - Escribe las fórmulas en celdas vacías para calcular media, mediana y moda.
    - Interpreta los resultados obtenidos.
- **Gráficas de tendencia central:**
  - Crear gráficos de barras y líneas.
  - **Paso a Paso:**
    - Selecciona los datos calculados (media, mediana y moda).
    - Ve a la pestaña "Insertar" > "Gráficos" y elige "Gráfico de columnas" o "Gráfico de líneas".
    - Personaliza el gráfico según sea necesario.

**3:30 PM - 4:30 PM:** Análisis de Correlación y Gráficas de Cajas y Bigotes
- **Análisis de Correlación:**
  - **Correlación:** `=CORREL(array1, array2)`
  - **Paso a Paso:**
    - Selecciona dos rangos de datos para comparar.
    - Usa la fórmula `=CORREL(array1, array2)` para calcular el coeficiente de correlación.
    - Interpreta el resultado (1 indica una correlación perfecta positiva, -1 indica una correlación perfecta negativa, 0 indica ninguna correlación).
- **Gráficas de Cajas y Bigotes:**
  - **Paso a Paso:**
    - Selecciona los datos que quieres visualizar.
    - Ve a "Insertar" > "Gráficos" > "Cajas y Bigotes".
    - Interpreta el gráfico, identificando la mediana, cuartiles y posibles valores atípicos.

**4:30 PM - 5:30 PM:** Regresión Lineal y Análisis de Tendencias
- **Regresión Lineal:**
  - **Ecuación de la recta:** `=LINEST(y_conocido, x_conocido)`
  - **Paso a Paso:**
    - Selecciona los datos de la variable dependiente (y) y la independiente (x).
    - Usa la fórmula `=LINEST(y_conocido, x_conocido)` para calcular los coeficientes.
    - Inserta un gráfico de dispersión y añade una línea de tendencia.
- **Análisis de Tendencias:**
  - **Línea de tendencia:**
    - Ve a "Insertar" > "Gráficos" > "Línea de tendencia".
    - El

ige el tipo de tendencia que mejor se ajuste a los datos (lineal, exponencial, polinómica).
    - Interpreta la línea de tendencia en el contexto de los datos.

**5:30 PM - 6:30 PM:** Elaboración de Tabla de Frecuencias y Histogramas
- **Tabla de Frecuencias:**
  - **Fórmulas:** `=FRECUENCIA(rango_de_datos, intervalos)`, `=CONTAR.SI(rango, criterio)`
  - **Paso a Paso:**
    - Define los intervalos para la variable de interés (e.g., edades).
    - Usa `=FRECUENCIA` para contar el número de observaciones en cada intervalo.
    - Usa `=CONTAR.SI` para contar frecuencias específicas.
    - Organiza los resultados en una tabla.
- **Histogramas:**
  - **Paso a Paso:**
    - Selecciona los datos de la tabla de frecuencias.
    - Ve a "Insertar" > "Gráficos" > "Histograma".
    - Personaliza el histograma y analiza la distribución de los datos.

**6:30 PM - 7:00 PM:** Introducción al análisis exploratorio de datos utilizando Power BI
- **Instalación y Configuración:**
  - **Paso a Paso:**
    - Descarga Power BI desde la página oficial.
    - Sigue las instrucciones de instalación y configura tu primera conexión de datos.
- **Primer Tablero Interactivo:**
  - **Paso a Paso:**
    - Importa el dataset de Titanic.
    - Crea visualizaciones básicas como gráficos de barras, de líneas y de sectores.
    - Publica y comparte el tablero con tus compañeros.

**Conclusiones y Resumen de la Clase**
- Discusión sobre la importancia del análisis exploratorio de datos.
- Preguntas y respuestas.
- Próximos pasos en el aprendizaje del análisis de datos.
