### Plan de Clase: Análisis Gráfico del Dataset del Titanic

**Duración:** 3 horas

**Objetivo:**
Al finalizar la clase, los estudiantes serán capaces de visualizar y analizar datos utilizando gráficos en Excel o Google Sheets. Serán capaces de identificar patrones y extraer conclusiones significativas del dataset del Titanic.

#### Agenda de la Clase

1. **Introducción y Preparativos (30 minutos)**
   - Introducción al análisis gráfico.
   - Breve revisión del dataset del Titanic.
   - Explicación de tipos de gráficos: histogramas, gráficos de barras, gráficos de dispersión, boxplots y gráficos de tarta.
   - Importancia de manejar datos vacíos.
   - Instalación y preparación del software (Excel o Google Sheets).

2. **Crear y Analizar Histogramas (30 minutos)**
   - Crear histogramas para la variable `Age`.
   - Análisis de la distribución de edades.
   - Crear histogramas para la variable `Fare`.
   - Análisis de la distribución de tarifas.

3. **Gráficos de Barras: Supervivencia por Género y Clase (30 minutos)**
   - Crear gráficos de barras para analizar la supervivencia por género.
   - Análisis y discusión de resultados.
   - Crear gráficos de barras para analizar la supervivencia por clase.
   - Análisis y discusión de resultados.

4. **Gráfico de Dispersión: Tarifa vs. Edad (30 minutos)**
   - Crear gráficos de dispersión para analizar la relación entre tarifa y edad.
   - Análisis y discusión de resultados.

5. **Boxplots y Gráficos de Tarta (30 minutos)**
   - Crear boxplots para `Age` y `Fare` por `Pclass`.
   - Análisis de distribución y detección de outliers.
   - Crear gráficos de tarta para la proporción de supervivencia general y por género/clase.
   - Análisis de proporciones y distribución.

6. **Manejo de Datos Vacíos y Conclusiones Generales (30 minutos)**
   - Estrategias para manejar datos vacíos.
   - Aplicación de métodos para rellenar datos vacíos.
   - Resumen de conclusiones obtenidas de los análisis gráficos.
   - Discusión final y preguntas.

### Detalles del Plan de Clase

#### 1. Introducción y Preparativos (30 minutos)

- **Introducción al Análisis Gráfico:**
  - Explica la importancia de los gráficos en la visualización de datos.
  - Muestra ejemplos de gráficos que se utilizarán.

- **Revisión del Dataset del Titanic:**
  - Breve descripción del dataset y sus principales columnas (`Survived`, `Pclass`, `Sex`, `Age`, `Fare`).

- **Tipos de Gráficos:**
  - Explica qué son los histogramas, gráficos de barras, gráficos de dispersión, boxplots y gráficos de tarta.
  - Muestra ejemplos de cada tipo de gráfico.

- **Manejo de Datos Vacíos:**
  - Discute la importancia de manejar datos vacíos.
  - Presenta diferentes estrategias para tratar datos vacíos.

- **Preparación del Software:**
  - Ayuda a los estudiantes a asegurarse de que tienen acceso a Excel o Google Sheets.
  - Verifica que el dataset del Titanic está cargado correctamente.

#### 2. Crear y Analizar Histogramas (30 minutos)

- **Crear Histogramas para la Variable `Age`:**
  - Guía paso a paso sobre cómo crear un histograma en Excel o Google Sheets.
  - Análisis de la distribución de edades (e.g., mayoría de pasajeros en el rango de 20-40 años).

- **Crear Histogramas para la Variable `Fare`:**
  - Guía paso a paso sobre cómo crear un histograma para `Fare`.
  - Análisis de la distribución de tarifas (e.g., muchas tarifas bajas, pocas tarifas muy altas).

#### 3. Gráficos de Barras: Supervivencia por Género y Clase (30 minutos)

- **Gráfico de Barras para Supervivencia por Género:**
  - Crear una tabla resumen con la supervivencia por género.
  - Crear el gráfico de barras.
  - Análisis y discusión (e.g., mayor supervivencia en mujeres).

- **Gráfico de Barras para Supervivencia por Clase:**
  - Crear una tabla resumen con la supervivencia por clase.
  - Crear el gráfico de barras.
  - Análisis y discusión (e.g., mayor supervivencia en primera clase).

#### 4. Gráfico de Dispersión: Tarifa vs. Edad (30 minutos)

- **Crear Gráfico de Dispersión para Tarifa vs. Edad:**
  - Guía paso a paso sobre cómo crear un gráfico de dispersión.
  - Análisis de la relación (e.g., no hay una relación clara entre tarifa y edad).

- **Exploración de Otros Gráficos de Dispersión:**
  - Sugerencias para otros gráficos de dispersión que se pueden explorar (e.g., `Age` vs. `Survived`).

#### 5. Boxplots y Gráficos de Tarta (30 minutos)

- **Crear Boxplots para Edad por Clase:**
  - Guía paso a paso sobre cómo crear un boxplot en Excel o Google Sheets.
  - Análisis de la distribución de edades por clase (e.g., medianas, cuartiles, outliers).

- **Crear Boxplots para Tarifa por Clase:**
  - Guía paso a paso sobre cómo crear un boxplot para `Fare` por `Pclass`.
  - Análisis de la distribución de tarifas por clase (e.g., medianas, cuartiles, outliers).

- **Crear Gráficos de Tarta:**
  - Guía paso a paso sobre cómo crear un gráfico de tarta para la proporción de supervivencia general y por género/clase.
  - Análisis de proporciones y distribución (e.g., mayor supervivencia en mujeres y en primera clase).

#### 6. Manejo de Datos Vacíos y Conclusiones Generales (30 minutos)

- **Estrategias para Manejar Datos Vacíos:**
  - Explicación de métodos como eliminación de filas/columnas, imputación con media/mediana/moda.
  - Aplicación práctica en el dataset del Titanic.

- **Resumen de Conclusiones:**
  - Revisión de las conclusiones obtenidas de los análisis gráficos.
  - Discusión final sobre qué información se puede extraer y cómo puede ser útil.

- **Preguntas y Respuestas:**
  - Espacio para que los estudiantes hagan preguntas y aclaren dudas.
  - Discusión sobre posibles proyectos futuros utilizando análisis gráfico.
